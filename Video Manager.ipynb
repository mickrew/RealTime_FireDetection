```python
import moviepy.editor as moviepy

def convert_video(pathVideo, source_format=".mov", dest_format=".mp4"):
  video = moviepy.VideoFileClip(pathVideo)
  new_pathVideo = pathVideo.replace(source_format, dest_format)
  video.write_videofile(new_pathVideo)
```

```python
import cv2
import os
def extract_frames(pathVideo, num_frames_per_second, pathFrames):

  video = cv2.VideoCapture(pathVideo)

  fps = video.get(cv2.CAP_PROP_FPS)
  print ( "Frames per second: {0}".format(fps))

  #num_frames_per_second = 2
  step = int(fps/num_frames_per_second)

  success,image = video.read()
  count = 0
  while success:
    if count % step == 0:
      id = str(int(count/step)).zfill(5)
      cv2.imwrite(pathFrames+"%s.jpg" % id, image)     # save frame as JPEG file      
    success,image = video.read()
    #print('Read a new frame: ', success)
    count += 1
  print("Saved " + str(int(count/step)) + " frames")
```
