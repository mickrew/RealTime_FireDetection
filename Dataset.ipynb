```python
def undersampling(samplingRate):
  #samplingRate is the number of images per second
  if os.path.isdir("training")==False: 
    os.mkdir("training")
    os.mkdir("training/Fire")
    os.mkdir("training/No_Fire")
  rate = int(29/samplingRate)
  
  for dir in os.listdir("Training"):
    path = "Training/"+ dir
    elems = os.listdir(path)
    elems.sort()
    for iteration,elem in enumerate(elems):
      if iteration%rate==0:
        shutil.move(path+"/" + elem, "training/"+dir)
```

```python
def resize_images(path):
  directories = os.listdir(path)
  for directory in directories:
    new_path = path+"/"+directory+"/"
    images = os.listdir(path+"/"+directory)
    for elem in tqdm.tqdm(images):
      try:
        im = Image.open(new_path + elem)
        im.thumbnail((508,254))
        
        width, height = im.size   # Get dimensions
        new_width, new_height = 254,254

        left = (width - new_width)/2
        top = (height - new_height)/2
        right = (width + new_width)/2
        bottom = (height + new_height)/2

        # Crop the center of the image
        im = im.crop((left, top, right, bottom))
        im.save(new_path + elem)
      except:
        os.remove(new_path + elem)
```

```python
resize_images("/content/fire_dataset")
```

```python
resize_images("/content/Fire-Detection")
```

```python
def moveFilesToFolder(source, dest):
    source_files = os.listdir(source)
    for file in tqdm.tqdm(source_files):
        shutil.move(source + file, dest)
```

```python
def rename_files(pathFolders):
  folders = os.listdir(pathFolders)
  folders.sort()
  idFolder = ["a", "b", "c"]
  for iteration, folder in enumerate(folders):
    files = os.listdir(os.path.join(pathFolders, folder))
    for file in files:
      newname = idFolder[iteration] + file
      os.rename(os.path.join(pathFolders, folder, file), os.path.join(pathFolders, folder, newname))
      shutil.move(os.path.join(pathFolders, folder, newname), "/content/trainingSet")

```

```python
rename_files("/content/To_Rename")
```
